```python
numerical_columns = df.select_dtypes(include=['int64', 'float64']).columns
categorical_columns = df.select_dtypes(include=['object']).columns

# You can further categorize numerical columns into discrete and continuous based on some criteria.
# For simplicity, let's say columns with less than 10 unique values are discrete, and others are continuous.
discrete_columns = [col for col in numerical_columns if df[col].nunique() < 10]
continuous_columns = [col for col in numerical_columns if col not in discrete_columns]

# Print the count for each category
print("Numerical columns (Count:", len(numerical_columns), "):")
print(numerical_columns)

print("\nCategorical columns (Count:", len(categorical_columns), "):")
print(categorical_columns)

print("\nDiscrete columns (Count:", len(discrete_columns), "):")
print(discrete_columns)

print("\nContinuous columns (Count:", len(continuous_columns), "):")
print(continuous_columns)
```

```python
# Calculate the percentage of missing values for each column
missing_percentage = df.isnull().mean() * 100

# Create a DataFrame to display the results
missing_info = pd.DataFrame({
    'Column': df.columns,
    'Missing Percentage': missing_percentage
})

# Print the DataFrame
print(missing_info)
```

```python
# Remove rows with missing values
df_cleaned_rows = df.dropna()

# Remove columns with missing values
df_cleaned_columns = df.dropna(axis=1)
```

```python
# Print the cleaned DataFrame after removing rows with missing values
print("DataFrame after removing rows with missing values:")
print(df_cleaned_rows.head())

# Print the cleaned DataFrame after removing columns with missing values
print("\nDataFrame after removing columns with missing values:")
print(df_cleaned_columns.head())
```

```python
# Print all columns for reference
print(df.columns)

# Replace 'numerical_column' and 'categorical_column' with actual column names
# Impute missing numerical values with mean
numerical_columns = ['age', 'wage_per_hour', 'working_week_per_year', 'gains', 'losses']
df[numerical_columns] = df[numerical_columns].fillna(df[numerical_columns].mean())

# Impute missing categorical values with mode
categorical_columns = ['gender', 'education', 'class', 'marital_status', 'race', 'is_hispanic', 'employment_commitment', 'unemployment_reason', 'employment_stat', 'is_labor_union', 'industry_code', 'industry_code_main', 'occupation_code', 'occupation_code_main', 'household_stat', 'household_summary', 'under_18_family', 'veterans_admin_questionnaire', 'vet_benefit', 'tax_status', 'stocks_status', 'citizenship', 'country_of_birth_own', 'country_of_birth_father', 'country_of_birth_mother', 'migration_code_change_in_msa', 'migration_prev_sunbelt', 'migration_code_move_within_reg', 'migration_code_change_in_reg', 'residence_1_year_ago', 'old_residence_reg', 'old_residence_state', 'importance_of_record', 'income_above_limit']
df[categorical_columns] = df[categorical_columns].fillna(df[categorical_columns].mode().iloc[0])
```

```python
from sklearn.impute import KNNImputer
from sklearn.preprocessing import LabelEncoder

# Separate numerical and categorical columns
numerical_columns = df.select_dtypes(include=['float64']).columns
categorical_columns = df.select_dtypes(include=['object']).columns

# Impute missing values for numerical columns using KNNImputer
imputer_numerical = KNNImputer(n_neighbors=2)
df[numerical_columns] = imputer_numerical.fit_transform(df[numerical_columns])

# Apply label encoding to categorical columns
label_encoder = LabelEncoder()
for col in categorical_columns:
    df[col] = label_encoder.fit_transform(df[col])

# Impute missing values for categorical columns using KNNImputer
imputer_categorical = KNNImputer(n_neighbors=2)
df[categorical_columns] = imputer_categorical.fit_transform(df[categorical_columns])

# Now, you can use the entire DataFrame for further analysis
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

for col in numerical_columns:
    plt.figure(figsize=(8, 5))
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribution of {col}')
    plt.show()
```

```python
# Bivariate Analysis: 
# Explore relationships between pairs of variables.
# Use scatter plots for numerical vs. numerical relationships.
```

```python
# Use crosstabs or stacked bar charts for categorical vs. categorical relationships
```

```python
for col in categorical_columns:
    ct = pd.crosstab(df[col], df['target_variable'])
    ct.plot(kind='bar', stacked=True, figsize=(10, 6))
    plt.title(f'{col} vs. Target Variable')
    plt.show()
```
